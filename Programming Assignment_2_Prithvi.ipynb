1. Write a Pythoin program to convert kilometers to miles?

```python
dist = float(input("Enter the distance in kilometer- "))
dist_miles = 1.6 * dist

print(f"The distance your is {dist} km and the distance in miles is {dist_miles} miles.")
```

2. Write a Python to convert Celsius to Fahrenheit?

```python
temp_cel = float(input("Enter the temperature in celsius; "))
temp_feh = (temp_cel * (9/5) + 32)

print(f"The temperature is {temp_cel} C and in Fahrenhiet is {temp_feh}")
```

3. Write a Python program to display calender?

```python
import calendar

year = int(input("Enter the year: "))
month = int(input("Enter the month: "))

cal = calendar.month(year, month)

print(cal)
```

4. Write a Python program to solve quadratic equation?

```python
a = int(input("Enter the coefficient of x^2: "))
b = int(input("Enter the coefficient of x: "))
c = int(input("Enter the value of the constant: "))

x = ((b*b -(4*a*c))**(1/2))
x_1 = round((-b+x)/(2*a),2)
x_2 = round((-b-x)/(2*a),2)

print(f"The quadratic equation is {a}x^2 + {b}x + {c} = 0, and it's solutions are {x_1} and {x_2}")
```
