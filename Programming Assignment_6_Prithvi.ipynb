1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
def fib(num):
    if num <= 1:
        return num
    else:
        return (fib(num-1) + fib(num-2))

term = int(input("Enter the term upto which fibancci sequwnce you want "))
if term <= 0:
    print("Plese enter a positive integer")  
else:
    print("Fibonacci sequence:")  
    for i in range(term):
        print(fib(i))
```

2. Write a Python Program to Find the Factorial of Number Using Recursion? 

```python
def factorial(x):
    if x == 1:
        return x
    else:
        return x * factorial(x-1)

num = int(input("Enter the number to find its factorial: "))
if num < 0:
    print("Factorial of negative number dosen't exist")
elif num == 0:
    print("Factorial of 0 is 1")
else:
    print("Factorial of", num, "is", factorial(num))
```

3. Write a Python Program to calculate your Body Mass Index?

```python
w = float(input("Enter your weight in kg: "))
h = float(input("Enter your height in meters-  "))

def bmi(w,h):
    bmi = w / (h*h)
    return bmi
bmi(w,h)
```
