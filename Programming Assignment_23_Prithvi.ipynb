1 .Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not. A number is symmetrical when it is the same as
its reverse.

```python
num = int(input("Is the entered number is symmetrical? "))

def is_symmetrical(num):
    temp = num
    r = 0
    while num>0:
        dig = num % 10
        r = (r*10) + dig
        num = num//10
    if temp == r:
        return True
    else:
        return False

is_symmetrical(num)        
```

2. Given a string of numbers separated by a comma and space, return the product of the numbers.

```python
s = input("Enter a number ")
def multiply_nums(s):
    l = []
    for i in s.split(","):
        l.append(i)
    y = 1
    for num in l:
        n = int(num)
        y = n * y
    return(y)

multiply_nums(s)
```

3. Create a function that squares every digit of a number.

```python
num = int(input("Enter the number: "))

def squares(num):
    l=[]
    for i in str(num):
        l.append(str(int(i)**2))
    return int(''.join(l))

squares(num)
```

5. Create a function that returns the mean of all digits.

```python
num = int(input("Enter the number: "))

def mean(num):
    l=[]
    a = 0
    for i in str(num):
        a += int(i)
        m = a/(len(str(num)))
    l.append(str(m))
        
    return round(float(''.join(l)),0)

mean(num)
```
