1. Write a Python Program to Find the Factorial of a Number?

```python
num = int(input("Enter the number to find its factorial: "))
def factorial(num):
    fact = 1
    if num < 0:
        print("Factorial of negative number dosen't exist")
    elif num == 0:
        print("Factorial of 0 is 1")
    else:
        for i in range(1, num+1):
            fact = fact * i
        return (f"Factorial of {num} is {fact}")
    
factorial(num)
```

2. Write a Python Program to Display the multiplication Table?

```python
num = int(input("Enter the number for which you want the multiple table: "))
upper_limit = int(input("Upto which you want the table to run like 10, 15, 20 or like that: "))

multiple = num
for i in range(1, upper_limit+1):
    multiple = num * i
    print(f"{num} X {i} = {multiple}")

```

3. Write a Python Program to Print the Fiboncci sequence?

```python
term = int(input("Enter the term upto which fibancci sequwnce you want"))
a = 0
b = 1

if term <= 0:
    print("Enter only positive number greater than 2")
else:
    print(a,b,end = " ")
    for i in range(2, term):
        i = a + b
        print(i, end = " ")
        a = b
        b = i
```

4. Write a Python Program to Check Armstrong Number?

```python
num = int(input("Enter a numbetr: "))

def armstong(num):
    sum = 0

    t = num

    while t>0:
        digit = t % 10
        sum += digit ** len(str(num))
        t //= 10

    if num == sum:
        return("The number is an armstrong number")
    else:
        return("Not an armstrong number")

armstong(num)
```

6. Write a Python Program to Find the Sum of Natural Number?

```python
upper_limit = int(input("Enter the upper limit: "))

def nat_sum(upper_limit):
    s = 0
    if upper_limit > 0:
        for i in range(upper_limit+1):
            s = s + i
        return s
    else:
        return ("Please enter a positive number")

nat_sum(upper_limit)
```
