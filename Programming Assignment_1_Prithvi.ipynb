3. Write a Python program to find the area of a triangle?

```python
a = int(input("Enter first side of triangel - "))
b = int(input("Enter second side of triangel - "))
c = int(input("Enter third side of triangel - "))
s = ((a + b + c)/2) 
"""'s' is the semi perimeter of the triangle, and is requierd to find the area according to Heron's Formula"""
x = (s*(s-a)*(s-b)*(s-c))
area = round(x**(1/2), 2)

print(f"The area of trinagle having sides {a}, {b} and {c} is {area} units")
```
