1. Write a Python Program to Find LCM?

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
def lcm(num1, num2):
    if(num1 > num2):
        min1 = num1
    else:
        min1 = num2
    while(True):
        if(min1 % num1 == 0 and min1 % num2 == 0):
            return (f"LCM is: {min1}")
            break
        min1 = min1+1
        
        
lcm(num1, num2)
```

2. Write a Oython Program to Find HCF?

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
def hcf(a, b):
    if(a > b):
        small = b
    else:
        small = a
    for i in range(1, small+1):
        if((a % i == 0) and (b % i == 0)):
            hcf = i
    return hcf
            
print("The HCF is ", hcf(a, b))
```
